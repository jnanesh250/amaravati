```python
# Qiskit 1.x — BB84 demo: 10 trials, random bases at Alice & Bob
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator

# ===== User input =====
user_bits_str = input("Enter 10 bits (e.g., 1011001101): ").strip()
assert len(user_bits_str) == 10 and set(user_bits_str) <= {"0","1"}, "Please enter exactly 10 bits."
alice_bits = np.array([int(b) for b in user_bits_str], dtype=int)

# Random bases: 0=Z, 1=X
alice_bases = np.random.choice([0,1], size=10)
bob_bases   = np.random.choice([0,1], size=10)

# ===== Registers =====
qr_data = QuantumRegister(1, "q")      # data qubit reused across trials
qr_a    = QuantumRegister(1, "qa")     # ancilla for Alice basis flag
qr_b    = QuantumRegister(1, "qb")     # ancilla for Bob basis flag
c_out   = ClassicalRegister(10, "m")   # Bob's measured bits
cA      = ClassicalRegister(1, "a")    # Alice basis flag
cB      = ClassicalRegister(1, "b")    # Bob basis flag
qc = QuantumCircuit(qr_data, qr_a, qr_b, c_out, cA, cB)

def set_flag_and_measure(q_anc, flag_val, cflag):
    if flag_val == 1:
        qc.x(q_anc)         # prepare |1> if flag=1
    qc.measure(q_anc, cflag)
    qc.reset(q_anc)         # reset ancilla for reuse

# ===== Build 10 trials =====
for i in range(10):
    # Encode random bases into classical flags
    set_flag_and_measure(qr_a[0], int(alice_bases[i]), cA[0])
    set_flag_and_measure(qr_b[0], int(bob_bases[i]),   cB[0])

    # Alice prepares bit
    if alice_bits[i] == 1:
        qc.x(qr_data[0])                # |1>, else stays |0>

    # If Alice basis = X, apply H
    with qc.if_test((cA, 1)):
        qc.h(qr_data[0])

    # If Bob basis = X, rotate by H before measurement
    with qc.if_test((cB, 1)):
        qc.h(qr_data[0])

    qc.measure(qr_data[0], c_out[i])    # Bob’s outcome
    qc.reset(qr_data[0])                # reuse qubit

# ===== Run =====
sim = AerSimulator()
tqc = transpile(qc, sim)
result = sim.run(tqc, shots=1, memory=True).result()

# ===== Extract outcomes =====
memory = result.get_memory(0)           # list of 1 string since shots=1
raw_str = memory[0]                     # full string for all regs: 'ab m9..m0'
m_bits = raw_str[-10:]                  # last 10 bits = register "m"
bob_meas = list(m_bits[::-1])           # flip order to align with trial index 0..9

# ===== Print table =====
print("\nTrial | AliceBit | AliceBasis | BobBasis | BobMeasured")
print("------+----------+------------+----------+------------")
for i in range(10):
    a_bit = int(alice_bits[i])
    a_bas = 'X' if alice_bases[i] == 1 else 'Z'
    b_bas = 'X' if bob_bases[i]   == 1 else 'Z'
    b_mea = bob_meas[i]
    print(f"{i:5d} | {a_bit:9d} | {a_bas:10s} | {b_bas:8s} | {b_mea:10s}")

matched = [i for i in range(10) if alice_bases[i] == bob_bases[i]]
print("\nMatched-basis indices:", matched)
```

```python
# === BB84 on IBM Quantum (10 jobs, 1 trial per job) ===
# Emulates BB84 steps on a single device; true QKD needs photon-based quantum channels.
# Uses Sampler V2 and extracts both counts and per-shot bitstrings from the named register "c".

# ONE-TIME on a secure machine (do NOT print your token):
from qiskit_ibm_runtime import QiskitRuntimeService
QiskitRuntimeService.save_account(
    channel="ibm_quantum_platform",
    token="azdphB1IM7YDuJxfN1DUeRRapIcYeKU2rYIQtk-Vfs7B",
    overwrite=True
)

from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2 as Sampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.visualization import plot_histogram
import numpy as np

# ===== Settings =====
INSTANCE = "crn:v1:bluemix:public:quantum-computing:us-east:a/c4c05eaac37c47eea00a48dd0588c1c1:dc60b26e-9ba6-4341-af62-01538cf352b7::" 
N_TRIALS = 10
SHOTS_PER_JOB = 1  # keep 1 to mirror the prior “one-shot per trial” demo; increase (e.g., 128) for histograms

# ===== Connect & pick backend =====
service = QiskitRuntimeService(
    channel="ibm_quantum_platform",
    instance=INSTANCE
)
backend = service.least_busy(simulator=False, operational=True)
print("Using backend:", backend.name)

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
sampler = Sampler(backend)

# ===== Helper: build one-trial BB84 circuit (no Eve) =====
def build_bb84_trial_circuit(alice_bit: int, alice_basis: int, bob_basis: int) -> QuantumCircuit:
    # alice_basis: 0=Z, 1=X ; bob_basis: 0=Z, 1=X
    q = QuantumRegister(1, "q")
    c = ClassicalRegister(1, "c")  # <-- named "c" so pub.data.c.get_* works
    qc = QuantumCircuit(q, c)

    # Alice prepares the bit in her basis
    if alice_bit == 1:
        qc.x(q[0])           # |1>, else stays |0>
    if alice_basis == 1:
        qc.h(q[0])           # X-basis prep: |+> or |->

    # Bob chooses his measurement basis (H -> measure Z == measure in X)
    if bob_basis == 1:
        qc.h(q[0])

    qc.measure(q[0], c[0])
    return qc

# ===== Helper to extract counts + per-shot memory from a job later =====
def counts_and_bits_from_job(job_id: str, creg="c"):
    job = service.job(job_id)
    res = job.result()
    pub = res[0]
    # Try both attribute and dict styles
    for getter in (
        lambda k: (getattr(pub.data, k).get_counts(), getattr(pub.data, k).get_bitstrings()),
        lambda k: (pub.data[k].get_counts(), pub.data[k].get_bitstrings()),
    ):
        try:
            return getter(creg)
        except Exception:
            pass
    # Fallback to the first key if names differ
    k0 = list(pub.data.keys())[0]
    return pub.data[k0].get_counts(), pub.data[k0].get_bitstrings()

# ===== User input & random bases (same spirit as earlier demo) =====
user_bits_str = input(f"Enter {N_TRIALS} bits (e.g., 1011001101): ").strip()
assert len(user_bits_str) == N_TRIALS and set(user_bits_str) <= {"0","1"}, f"Please enter exactly {N_TRIALS} bits."
alice_bits = np.array([int(b) for b in user_bits_str], dtype=int)

alice_bases = np.random.choice([0, 1], size=N_TRIALS)  # 0=Z, 1=X
bob_bases   = np.random.choice([0, 1], size=N_TRIALS)  # 0=Z, 1=X

# ===== Submit 10 jobs (one circuit per job) =====
job_ids = []
for i in range(N_TRIALS):
    qc = build_bb84_trial_circuit(alice_bits[i], alice_bases[i], bob_bases[i])
    isa_circ = pm.run(qc)
    job = sampler.run([isa_circ], shots=SHOTS_PER_JOB)
    job_ids.append(job.job_id())
    print(f"[Submitted] Trial {i} -> job_id = {job_ids[-1]}")

print("\nAll job IDs:")
for i, jid in enumerate(job_ids):
    print(f"  Trial {i}: {jid}")

# ===== Collect results (waits for each job) =====
print("\nTrial | AliceBit | AliceBasis | BobBasis | BobMeasured | job_id")
print("------+----------+------------+----------+-------------+--------------------------")
bob_measured = []
for i, jid in enumerate(job_ids):
    counts, bits = counts_and_bits_from_job(jid, creg="c")
    # bits is a list of per-shot strings; with SHOTS_PER_JOB=1 -> e.g., ['0'] or ['1']
    mea = bits[0] if len(bits) else "?"
    bob_measured.append(mea)
    a_bas = 'X' if alice_bases[i] == 1 else 'Z'
    b_bas = 'X' if bob_bases[i]   == 1 else 'Z'
    print(f"{i:5d} | {alice_bits[i]:9d} | {a_bas:10s} | {b_bas:8s} | {mea:11s} | {jid}")

matched = [i for i in range(N_TRIALS) if alice_bases[i] == bob_bases[i]]
print("\nMatched-basis indices:", matched)

# Optional: quick sanity histogram if you increased SHOTS_PER_JOB
# from qiskit.visualization import plot_histogram
# plot_histogram(counts, title=f"Trial {i} (last) — counts on {backend.name}")
```

# QKD : BB84 (Eve)

```python
# BB84 — Simulation with and without Eve (Intercept–Resend)
# Qiskit 1.x | AerSimulator
# Safer version: use get_counts() to retrieve 'm' register

import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator

# ---------- Config ----------
N_TRIALS = 100
SEED = 7

rng = np.random.default_rng(SEED)
alice_bits  = rng.integers(0, 2, size=N_TRIALS)
alice_bases = rng.integers(0, 2, size=N_TRIALS)   # 0=Z,1=X
bob_bases   = rng.integers(0, 2, size=N_TRIALS)
eve_bases   = rng.integers(0, 2, size=N_TRIALS)

# ---------- Helpers ----------
def build_bb84_no_eve(alice_bits, alice_bases, bob_bases):
    q  = QuantumRegister(1, "q")
    m  = ClassicalRegister(N_TRIALS, "m")
    qc = QuantumCircuit(q, m)
    for i in range(N_TRIALS):
        if alice_bits[i] == 1: qc.x(q[0])
        if alice_bases[i] == 1: qc.h(q[0])
        if bob_bases[i] == 1: qc.h(q[0])
        qc.measure(q[0], m[i])
        qc.reset(q[0])
    return qc

def build_bb84_with_eve(alice_bits, alice_bases, eve_bases, bob_bases):
    q  = QuantumRegister(1, "q")
    e  = ClassicalRegister(1, "e")
    m  = ClassicalRegister(N_TRIALS, "m")
    qc = QuantumCircuit(q, e, m)
    for i in range(N_TRIALS):
        if alice_bits[i] == 1: qc.x(q[0])
        if alice_bases[i] == 1: qc.h(q[0])
        if eve_bases[i] == 1: qc.h(q[0])
        qc.measure(q[0], e[0])
        qc.reset(q[0])
        with qc.if_test((e, 1)):
            qc.x(q[0])
        if eve_bases[i] == 1: qc.h(q[0])
        if bob_bases[i] == 1: qc.h(q[0])
        qc.measure(q[0], m[i])
        qc.reset(q[0])
    return qc

def run_and_extract_m_bits(qc, N_bits):
    sim = AerSimulator()
    tqc = transpile(qc, sim)
    result = sim.run(tqc, shots=1).result()  # 1 shot, returns counts
    counts = result.get_counts()
    # Only one outcome key since shots=1
    key = list(counts.keys())[0]
    # The 'm' register is the rightmost N_bits in the key
    m_token = key.replace(" ", "")[-N_bits:]
    m_bits_reversed = m_token[::-1]
    arr = np.array([1 if ch == '1' else 0 for ch in m_bits_reversed], dtype=int)
    return arr

def sift_and_qber(alice_bits, bob_bits, alice_bases, bob_bases):
    matched_idx = np.where(alice_bases == bob_bases)[0]
    a_sift = alice_bits[matched_idx]
    b_sift = bob_bits[matched_idx]
    matches = (a_sift == b_sift)
    L = len(a_sift)
    M = int(matches.sum())
    mm = L - M
    qber = (mm / L) * 100 if L else float('nan')
    return matched_idx, a_sift, b_sift, L, M, mm, qber

def bits_to_str(arr):
    return ''.join(map(str, arr.tolist())) if len(arr) else "(none)"

# ---------- Build & run ----------
qc_no_eve   = build_bb84_no_eve(alice_bits, alice_bases, bob_bases)
qc_with_eve = build_bb84_with_eve(alice_bits, alice_bases, eve_bases, bob_bases)

bob_no_eve   = run_and_extract_m_bits(qc_no_eve,   N_TRIALS)
bob_with_eve = run_and_extract_m_bits(qc_with_eve, N_TRIALS)

# ---------- Stats ----------
matched_idx, a_no_eve, b_no_eve, L0, M0, mm0, q0 = sift_and_qber(alice_bits, bob_no_eve,   alice_bases, bob_bases)
_,           a_eve,    b_eve,    L1, M1, mm1, q1 = sift_and_qber(alice_bits, bob_with_eve, alice_bases, bob_bases)

# ---------- Print ----------
print("=== No Eve ===")
print("Matched-basis positions:", matched_idx.tolist())
print("Alice sifted bits:", bits_to_str(a_no_eve))
print("Bob   sifted bits:", bits_to_str(b_no_eve))
print(f"Length={L0}, Matches={M0}, Mismatches={mm0}, QBER={q0:.2f}%\n")

print("=== With Eve (Intercept–Resend) ===")
print("Matched-basis positions:", matched_idx.tolist())
print("Alice sifted bits:", bits_to_str(a_eve))
print("Bob   sifted bits:", bits_to_str(b_eve))
print(f"Length={L1}, Matches={M1}, Mismatches={mm1}, QBER={q1:.2f}%")
```

```python
# BB84 — No Eve but with artificial noise
# Simulate hardware-like imperfections → small QBER (< 11%)

import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator, noise

# ---------- Config ----------
N_TRIALS = 100
SEED = 42
rng = np.random.default_rng(SEED)

alice_bits  = rng.integers(0, 2, size=N_TRIALS)
alice_bases = rng.integers(0, 2, size=N_TRIALS)
bob_bases   = rng.integers(0, 2, size=N_TRIALS)

# ---------- Build No Eve Circuit ----------
def build_bb84_no_eve(alice_bits, alice_bases, bob_bases):
    q  = QuantumRegister(1, "q")
    m  = ClassicalRegister(N_TRIALS, "m")
    qc = QuantumCircuit(q, m)
    for i in range(N_TRIALS):
        if alice_bits[i] == 1: qc.x(q[0])
        if alice_bases[i] == 1: qc.h(q[0])
        if bob_bases[i] == 1: qc.h(q[0])
        qc.measure(q[0], m[i])
        qc.reset(q[0])
    return qc

qc = build_bb84_no_eve(alice_bits, alice_bases, bob_bases)

# ---------- Add Noise Model ----------
# Depolarizing + Readout error to simulate real hardware
error_prob = 0.05   # 5% depolarizing noise
depol = noise.depolarizing_error(error_prob, 1)
ro_err = noise.errors.ReadoutError([[0.92, 0.08], [0.08, 0.92]])  # ~8% readout error

noise_model = noise.NoiseModel()
noise_model.add_all_qubit_quantum_error(depol, ['x','h'])
noise_model.add_all_qubit_readout_error(ro_err, ['measure'])

sim = AerSimulator(noise_model=noise_model)

# ---------- Run ----------
tqc = transpile(qc, sim)
result = sim.run(tqc, shots=1).result()
counts = result.get_counts()
key = list(counts.keys())[0].replace(" ","")
m_token = key[-N_TRIALS:]
bob_bits = np.array(list(m_token[::-1]), dtype=int)

# ---------- Sift + QBER ----------
matched_idx = np.where(alice_bases == bob_bases)[0]
a_sift = alice_bits[matched_idx]
b_sift = bob_bits[matched_idx]
matches = (a_sift == b_sift)
L = len(a_sift)
M = int(matches.sum())
mm = L - M
qber = (mm / L) * 100 if L else float('nan')

print("=== No Eve (with noise) ===")
print("Matched-basis positions:", matched_idx.tolist())
print("Alice sifted bits:", ''.join(map(str, a_sift.tolist())))
print("Bob   sifted bits:", ''.join(map(str, b_sift.tolist())))
print(f"Length={L}, Matches={M}, Mismatches={mm}, QBER={qber:.2f}%")

if qber < 11:
    print("\nQBER is below 11% → Alice & Bob can apply classical error correction (e.g. Cascade/LDPC) to reconcile keys.")
else:
    print("\nQBER too high → likely Eve or very noisy channel → discard key.")
```

```python
# ---------- Cascade Protocol (1-pass) ----------
def parity(bits):
    return sum(int(b) for b in bits) % 2

def cascade_pass(alice, bob, block_size=10):
    alice = list(alice)
    bob = list(bob)
    n = len(alice)
    corrections = []
    for start in range(0, n, block_size):
        end = min(start + block_size, n)
        a_block = alice[start:end]
        b_block = bob[start:end]
        if parity(a_block) != parity(b_block):
            low, high = 0, end - start
            while high - low > 1:
                mid = (low + high) // 2
                if parity(a_block[low:mid]) != parity(b_block[low:mid]):
                    high = mid
                else:
                    low = mid
            error_index = start + low
            bob[error_index] = '1' if bob[error_index] == '0' else '0'
            corrections.append(error_index)
    return "".join(bob), corrections

if qber < 11:
    alice_str = "".join(map(str, a_sift.tolist()))
    bob_str   = "".join(map(str, b_sift.tolist()))
    corrected_bob, corr = cascade_pass(alice_str, bob_str, block_size=10)
    print("\n=== Cascade Correction ===")
    print("Corrected Bob  :", corrected_bob)
    print("Corrections at :", corr)
    print("Final Match?   :", corrected_bob == alice_str)
```
